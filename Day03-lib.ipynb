- 취합하여 하나의 df로 완성, 시도명(인덱스), 도서관 갯수, 인구, 인구 1인당 도서관 수(pd.concat)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
kosis_data = pd.read_csv('인구_가구_및_주택_–_읍면동_연도_끝자리_0_5___시군구_그_외_연도__20230331214813.csv', encoding = 'cp949', header = 1)
#kosis_data = kosis_data[1:]
kosis_data
```

```python
pd_data = pd.read_csv('전국도서관표준데이터.csv', encoding = 'cp949')
pd_data.head()
```

```python
for i,n in enumerate(pd_data):
    print(i, n)
```

```python
sub_df = pd_data[['시도명', '도서관유형', '자료수(도서)', '열람좌석수']]
g1 = sub_df.groupby('도서관유형')
g1.count()
```

```python
g2 = sub_df.groupby(['시도명', '도서관유형']) # row 갯수를 count한다.
g2.count()
```

```python
g3 = g2.agg([np.mean, np.sum])
g3
```

```python
sub_df.describe()
```

```python
filter1 = sub_df['열람좌석수'] == 3989 # 3989 대신에 sub_df['열람좌석수'].max()해도 된다.
filter4 = sub_df['열람좌석수'] == sub_df['열람좌석수'].min()

print(sub_df[filter1])
print('')
print(sub_df[filter4])
```

```python
filter2 = sub_df['자료수(도서)'] == sub_df['자료수(도서)'].min()
filter3 = sub_df['자료수(도서)'] == sub_df['자료수(도서)'].max()

print('최소 :')
print(sub_df[filter2])
print('')
print('최대 :')
print(sub_df[filter3])
```

```python
g4 = sub_df.groupby('시도명').count()
kosis_data.set_index('행정구역별(읍면동)', inplace = True) # set_index()는 index를 내가 설정한 column으로 바꿔준다.
```

```python
print(g4)
print(kosis_data)
```

```python
df10 = pd.concat((kosis_data, g4), axis = 1)
df10['ratio'] = df10['총인구 (명)'].astype(np.float64)/df10['도서관유형'].astype(np.float64)
# ratio == 인구수 / 도서관 ==> 도서관 1개당 커버가능한 인구수 ==> 인구대비 도서관이 많은곳,작은곳
df10.insert(1, 'ratio2', df10['도서관유형'].astype(np.float64)/df10['총인구 (명)'].astype(np.float64))
```
