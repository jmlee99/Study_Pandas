#### IRIS
- sepal_len의 iqr, 이상치 upper/lower 값
- class 코드화: 'Iris-virginica', 'Iris-setosa', 'Iris-versicolor',  순으로 0, 1, 2 지정

```python
import pandas as pd

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
pd_iris = pd.read_csv(url, header=None)
pd_iris.head()
```

```python
pd_iris.columns = ['sepal_len', 'sepal_width', 'petal_len', 'petal_width', 'class']
pd_iris['class'].unique()

pd_iris
```

```python
qv = pd_iris['sepal_len'].quantile([0.25, 0.75])
qv = qv.to_numpy()
IQR = qv[1] - qv[0]

upper = qv[1] + 1.5*IQR
lower = qv[0] - 1.5*IQR

print('IQR:', IQR)
print('upper:', upper)
print('lower:', lower)

filter1 = pd_iris['sepal_len'] < lower
filter2 = pd_iris['sepal_len'] > upper


print(pd_iris['sepal_len'][filter1])
print(pd_iris['sepal_len'][filter2])
# 이상치 없음 --> 머신러닝돌리기 아주 좋은 dataset
```

```python
from pandas.api.types import CategoricalDtype

# class 그대로 순서
pd_iris['class'] = pd_iris['class'].astype('category')

pd_iris['cat_code'] = pd_iris['class'].cat.codes


# 내가 원하는 순서
area_name = ['Iris-virginica', 'Iris-setosa', 'Iris-versicolor']

area_cat = CategoricalDtype(categories=area_name, ordered = True)

pd_iris['class'] = pd_iris['class'].astype(area_cat)
pd_iris['class_cat_code'] = pd_iris['class'].cat.codes
```
