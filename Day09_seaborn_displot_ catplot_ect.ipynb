```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
pd_pen = sns.load_dataset('penguins')
pd_pen.head() # 종 어떤섬 부리길이 부리깊이 날개길이 몸무게 성별
```

```python
print(pd_pen['bill_length_mm'].describe())

sns.displot(data = pd_pen, x = 'bill_length_mm', bins = 10, hue='species', multiple = 'stack') 

# multiple에는 dodge(피하다), stack(쌓다)
#연속된 값을 넣으면 몇개있는지 분포를 알려줌 displot의 default값은 hist값이다.
#sns.displot(data = pd_pen, x = 'species') # 명목형 값을 넣으면 bar 차트가 된다.
```

```python
# kde

sns.displot(data = pd_pen, x = 'bill_length_mm', kind = 'kde', cut = 0)

# hist(분포)를 대충 선으로 그려주면 이렇다. - 예상값이다 ==> 정확한 것은 아니다.
# cut를 주면 데이터가 없는 부분에서 잘라준다. 원래는 예상해서 다 연결해준다.
```

```python
# ecdf

sns.displot(data = pd_pen, x = 'bill_length_mm', kind = 'ecdf', hue = 'species')

# ecdf : 예상되는 cdf(누적분포)를 그려준다. - 누적이 되어갈때 기울기를 주로 본다.
```

```python
#catplot - stripplot 
#여기서 부터는 x축에 명목형 값을 주어야 한다.

sns.catplot(data = pd_pen, x = 'species', y = 'bill_length_mm', jitter = 0.25, hue = 'sex') 

# 얼마나 많이 모여있는가를 확인하는 plot
# jitter은 폭을 넓혀달라는 것
```

```python
# swarmpplot

sns.catplot(data = pd_pen, x = 'species', y = 'bill_length_mm', kind = 'swarm', hue = 'sex')

# 옆으로 갯수를 count하는 것
```

```python
# boxplot

sns.catplot(data = pd_pen, x = 'species', y = 'bill_length_mm', kind = 'box', hue = 'sex')

# describe를 시각화 시킨것이다.
```

```python
# violinplot

sns.catplot(data = pd_pen, x = 'species', y = 'bill_length_mm', kind = 'violin', hue = 'sex')

# violin의 곡선은 kde(예상값) 분포이다 그것을 좌우대칭 시켜둔 것이다.
# 가운데 들어가있는 것은 boxplot이다.
```

```python
# pointplot 
sns.catplot(data = pd_pen, x = 'species', y = 'bill_length_mm', kind = 'point')

#marker 은 min값을 서로 연락해둔 것이다.
```

```python
#barplot

sns.catplot(data = pd_pen, x = 'species', y = 'bill_length_mm', kind = 'bar')
```

```python
# boxenplot

sns.catplot(data = pd_pen, x = 'species', y = 'bill_length_mm', kind = 'boxen')
```

```python
sns.pairplot(data = pd_pen, hue = 'species') 

# x 축 y 축에는 연속형만 깔려있다.
# 상관도를 봐야한다. 예측모델을 만들기 위해 (3,4) or (4,3)
# 그룹으로 구분되는 것을 찾아야한다. (2,3) or (3,2) 이런것들에서 또 주황이랑 파랑 분류되는 것은 있나? 계속 찾아가야한다.
# (3,4) or (4,3)에서는 종으로는 구분이 안되지만 상관도는 있구나 
```

```python
# x축 y축 둘 다 명목형 values이다.
# 데이터 형식은 df2 처럼 index도 명목형, col도 명목형이 되어야한다.

print(pd_pen)

g1 = pd_pen.groupby(['species', 'island'])
df1 = g1.mean().reset_index()
df2 = df1.pivot('species', 'island', 'body_mass_g').fillna(0.0)
# index에는 species를 깔아주고, y축에는 island를 깔아주고 
# index와 columns에 따라 values값이 하나씩만 가져야한다.
# 옆에 색깔을 보고 판단한다. count가 많을땐 밝은색, count가 적을 땐 어두운색

sns.heatmap(data = df2)
```

```python
# 다시 한 번 해보자

# heatmap은 주로 categories values가 많을 때 쓴다. (명목형 값이 많을 때 쓴다.)

g1 = pd_pen.groupby(['species', 'island'])
df5 = g1.count().reset_index()
df6 = df5.pivot('species', 'island', 'body_mass_g').fillna(0.0)


sns.heatmap(data  = df6)
#Biscoe섬에는 Gentoo가 가장 많다.
```
