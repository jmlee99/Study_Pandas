#### 전국자동차정비업체표준데이터
- 지자체별 (광역시도) 자동차 정비업체 갯수
- 지자체별 (광역시도) 자동차 정비업체, 영업상태별 갯수 
- (영업상태를 column의 multi-index)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def check_df(df):
    #print(df.head())
    for i, col_name in enumerate(df):
        print(i, col_name, df[col_name].dtypes)
        if df[col_name].hasnans:
            print(df[col_name].isna().value_counts())
        print('')
        if len(df[col_name].unique())<10:
            print(df[col_name].value_counts())
        print('')
```

```python
filename = '전국자동차정비업체표준데이터.csv'
pd_data = pd.read_csv(filename, encoding = 'cp949')
pd_data.head()
```

```python
check_df(pd_data)
```

```python
sub_df = pd_data[['관리기관명', '영업상태']]


def check_area(df):
    
    return df.split()[0]

area_name = sub_df['관리기관명'].apply(check_area)

sub_df['area_name'] = area_name

df10 = sub_df.groupby(['area_name', '영업상태'])

df11 = df10.count()
```

```python
df12 = df11.unstack()
```

```python
df12.columns = ['영업', '휴업', '폐업']

df12
```

```python
sub_df2 = pd_data[['소재지도로명주소', '소재지지번주소']]
filter1 = sub_df2['소재지도로명주소'].isna() # 빈칸인지 확인

sub_df2['소재지도로명주소'][filter1] = sub_df2['소재지지번주소'][filter1]
```

```python
sub_df3 = sub_df2['소재지도로명주소'].apply(check_area)

sub_df['area'] = sub_df3

group_df = sub_df.groupby(['area', '영업상태'])
```

```python
tot = group_df.count().drop('area_name', axis = 1).unstack()

tot.columns = ['영업', '휴업', '폐업']

df3 = tot.replace(np.NaN, 0)
```

```python
df3.to_csv('지자체별 영업 상태.csv')

import pickle

with open('지자체별_영업상태.pickle', 'wb') as f:
    pickle.dump(df3, f)
    
with open('지자체별_영업상태.pickle', 'rb') as f:
    df4 = pickle.load(f)
```

```python
df4 = df4.reset_index()
```

```python
df4['area'] = df4['area'].astype('category')
```

```python
df4['area'].cat.codes # dtypes가 category일때만 사용할 수 있는 코드 ==> 머신러닝을 위한 코드
df4['area_cat_code'] = df4['area'].cat.codes 
#df4.dtypes
df4
```

```python
# 순서를 지정하기 위해 문자열을 숫자화 시켜주기위해 .cat.codes를 해준다.
# import line 필수
from pandas.api.types import CategoricalDtype

area_name_list = ['서울특별시','광주광역시','대구광역시','대전광역시','부산광역시','울산광역시','인천광역시','세종특별자치시','제주특별자치도',
                  '강원도','경기도','경상남도','경상북도','전라남도', '전라북도' '충청남도', '충청북도']

area_cat = CategoricalDtype(categories=area_name_list, ordered=True)
df4['area_cat2'] = df4['area'].astype(area_cat)
print(df4.dtypes)
df4['area_car2_code'] = df4['area_cat2'].cat.codes # sklearn lablencoder
```

```python
sub_df5 = pd_data[['사업등록일자', '면적']]
print(sub_df5)


sub_df5['면적'] = sub_df5['면적'].replace('2351.3.', '2351.3')
sub_df5['면적'] = sub_df5['면적'].astype(np.float64)
sub_df5.dtypes
```

```python
#sub_df5['사업등록일자'].apply(lambda x: x.split('-'))
sub_df5['사업등록일자'] = sub_df5['사업등록일자'].astype('datetime64') # dateime64 --> 날짜를 받아오는 datatype
#sub_df5.dtypes
sub_df5['년'] = sub_df5['사업등록일자'].dt.year
sub_df5['월'] = sub_df5['사업등록일자'].dt.month 
sub_df5['일'] = sub_df5['사업등록일자'].dt.day
sub_df5
```

```python
sub_df5['면적'].describe() # 25% == Q1, 50%(중앙값) == Q2, 75% == Q3
                           # IQR = Q3 - Q1 : 이상치를 알아보는 값
```

```python
# 이상치 알아보는 법

qv = sub_df5['면적'].quantile([0.25, 0.75])
qv = qv.to_numpy()
print(qv)
IQR = qv[1] - qv[0]
print(IQR)

upper_bound = qv[1] + 1.5*IQR
lower_bound = qv[0] - 1.5*IQR
print(upper_bound)
print(lower_bound)

#이상치, upper_bound보다 큰 것, lower_bound보다 작은 것 == 이상치
filter1 = sub_df5['면적']<lower_bound
sub_df5['면적'][filter1]

filter2 = sub_df5['면적']>upper_bound
sub_df5.loc[filter2,:] # 이상치에 대해서만 분리를 해서 할려면
```
