```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 리눅스 한글 설정
plt.rc('font', family = 'NanumGothic')
```

```python
file = '전국신호등표준데이터.csv'
pd_data = pd.read_csv(file, encoding = 'cp949')
pd_data.head()
```

```python
xs = pd_data['시도명'].value_counts().index
ys = pd_data['시도명'].value_counts()

fig, axe = plt.subplots(figsize = (8,6))
# plt.style.use('_mpl-gallery')
axe.barh(xs, ys, color = ['b', 'r', 'g'], label=xs)
#axe.legend(['경기도', '대전관역시'])
axe.legend(ncols = 2)
#axe.set_xticklabels(xs, rotation=45)
```

```python
x = pd_data['위도']
y = pd_data['경도']

fig, axe = plt.subplots(figsize = (6,8))
axe.scatter(y, x, s=10, c='green', marker = '^')
axe.set_xticklabels(list('abcdefghij'))
```

```python
fig, axe = plt.subplots(figsize = (8,6))
axe.plot(xs, ys, 'g.--', label = 'myline1') # 녹색 점선
axe.set_xticklabels(xs, rotation = 45)
axe.legend(loc = 'center right')
```
