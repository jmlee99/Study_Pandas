```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0, 50, 100)
ys = np.sin(xs)

# 우리가 보는 fig나 axe를 artist라고 한다.

fig, axe = plt.subplots()
axe.plot(xs, ys)

fig.set_facecolor('red')
fig.set_edgecolor('blue')
fig.suptitle('Figure1')
fig.supxlabel('Figure_X')
fig.supylabel('Figure_Y')
fig.text(0.5, 0.7, 'mytext')

axe.set_facecolor('yellow')
axe.set_title('axe_title')
axe.set_xlabel('axe_x')
axe.set_ylabel('axe_y')
axe.set_xlim(0, 50)
axe.set_ylim(-2, 2)
axe.set_xticks(np.linspace(0, 30, 5)) # 0 ~ 30 까지 5개를 찍어줘
axe.set_xticklabels(list('abcde'))    # 그 곳에 이름을 abcde로 줘

axe.set(facecolor = 'violet', title = 'axe_title2') # 이렇게 한번에 줄 수도 있다.
```

```python
# 하나의 figure에 여러개의 그래프를 그리는 법

fig, axes = plt.subplots(2, 3, figsize = (12,8)) 
# 2행 3열 (row = 2, columns = 3), 여러개를 뜻하는 axes를 해준다.
# figsize로 figure를 조정 == figsize = (폭<가로>, 길이<세로>)

print(axes) # ndarray로 되어 있다. == index를 가진다.


#axes[0, 0].plot(xs, ys)
axes[1, 1].plot(xs, ys)
```

```python
#하나의 figure안에 figure를 하나를 더 넣는 법 (좌표활용)

fig1 = plt.figure(figsize = (12,8))

fig1.set_facecolor('red')
a = fig1.add_axes([0,0,1,1])
b = fig1.add_axes([0.3,0,0.5,0.5])

a.plot(xs, ys)
a.set_facecolor('yellow')
b.set_facecolor('violet')
b.plot(xs, ys)
```

```python
# 하나의 figure에 여러개의 figure 추가하기 (구역을 나누는 개념)

# add_subplot(axis0, axis1, index)
fig2 = plt.figure()
# (2,1,1) -> axis0으로 두개로 나누고, (axis로는 하나로 나눠 = 쪼개지마), 그리고 index를 불러준다.
axe1 = fig2.add_subplot(2,1,1)
axe2 = fig2.add_subplot(2,2,3)
axe3 = fig2.add_subplot(2,2,4)

axe2.plot(xs, ys)
axe3.set_facecolor('red')
```
