```python
pd_tips = sns.load_dataset('tips') # seaborn load_dataset은 git에서 들고오자 (구글 검색)
pd_tips.head() # total_bill, tip = 연속적인 값들이다.
               # sex, smoker, day, time, size = 명목적인 값들이다.
               # but, size는 숫자이지만 정해진 값들이기 때문에 명목적인 값으로 들어간다.
```

```python
# hue, style, size == (색깔, maker, 크기) <-- categorical var.(주로 명목적인 값을 준다.)

sns.relplot(data = pd_tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker', size='size')
# relplot는 x,y둘다 연속적인 값이여야한다.

# figure를 col, row로 짜를 수도 있다.
sns.relplot(data = pd_tips, x = 'total_bill', y = 'tip', col = 'sex', row = 'smoker')
```

```python
#relplot에서 다른 plot를 그리고 싶다면 kind를 쓰면 된다.

sns.relplot(data = pd_tips, x = 'total_bill', y = 'tip', kind = 'line')
```
